1.4. 자율주행

 1.4.1  웨이모
        알파벳의 자회사 자율주행 기술 개발 회사
        자율 주행 자동차는 운전에 필요한 다양한 정보를 얻은 후 이를 
        해석해 의사결정을 내린다.
        GPS (위성위치확인시스템)를 통해 현재 위치와 목적지를 끊임없이 
        비교하면서 원하는 방향으로 핸들을 돌린다.
        목적지를 설정한 후 규정된 지점(웨이포인트ㆍWaypoint)만 지나면
         자동운전되는 항공기와 같은 원리다. 
        여기에 레이다ㆍ카메라ㆍ레이저 스캐너가 도로의 다양한 정보
        (주변 차량ㆍ사물사람•신호)를 확보한다.
        GPS가 조향장치 개념이라면 이 장비들은 사물탐지ㆍ충돌방지 장치다. 
        이렇게 수집된 데이터는 구글 컴퓨터가 종합ㆍ분석해 방향조작, 가ㆍ감속,
        정지 등 운전에 필요한 최종의사결정을 내린다.
        2019년 웨이모는 미국 캘리포니아주에서 시험적으로 승객을 허가 받았으며
        2016년 애리조나주 피닉스 도심부에서도 일부 고객을 상대로 상업적 무인
         택시를 서비스 시험을 해봤습니다.
        자율주행은 개발 단계는 레벨3단계로써 특정 조건 구간에서 시스템이 주행을 
        담당하며 위험 상황시에 운전자가 개입할수있는 레벨이며
        대부분의 상황에서 자율주행이 가능한 4레밸은 2024~2025년으로 
        추정되고 있습니다.
